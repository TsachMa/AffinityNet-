```python
import openbabel

obConversion = openbabel.OBConversion()
obConversion.SetInAndOutFormats("pdb", "pdb")

mol = openbabel.OBMol()
obConversion.ReadFile(mol, "../../test_data/pdb/1a0q/1a0q_protein.pdb")

# Use the MMFF94 charge model
chargeModel = openbabel.OBChargeModel.FindType("mmff94")
chargeModel.ComputeCharges(mol)

for atom in openbabel.OBMolAtomIter(mol):
    print(f"Atom {atom.GetIdx()} charge: {atom.GetPartialCharge()}")

obConversion.WriteFile(mol, "../../test_data/pdb/1a0q/1a0q_protein_ob_charged.pdb")
```
